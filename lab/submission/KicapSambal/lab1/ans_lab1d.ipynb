# Lab 1d - Filtering and Sorting Data
#### The Chipotle Sales dataset is used for this exercise

## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

## Step 2. Import the dataset from the address given.
## Step 3. Assign it to a variable called chipo.


```python
# import dataset
chipo = pd.read_table('https://raw.githubusercontent.com/drshahizan/dataset/main/pandas/chipotle.tsv')
chipo.head(10);
```

```python
# info from dataset
chipo.info();
chipo.describe()
```

## Step 4. How many products cost more than $10.00?

```python
# Convert item_price to float
chipo['item_price'] = chipo['item_price'].astype(str)
chipo['item_price'] = chipo['item_price'].str.replace('$', '')
chipo['item_price'] = chipo['item_price'].astype(float)
```

```python
# Count product that cost more than $10.00
np.count_nonzero(chipo.item_price > 10.00)
```

## Step 5. What is the price of each item?
#### print a data frame with only two columns item_name and item_price

```python
# Identify the price for 1 item
chipo2 = chipo.groupby(by ="item_name")[["item_price"]].min().reset_index()
chipo2
```

## Step 6. Sort by the name of the item

```python
# Sort alphabetically
chipo.sort_values(by = "item_name").reset_index(drop=True)
```

## Step 7. What was the quantity of the most expensive item ordered?

```python
# Checking the most expensive item
chipo_expensive = np.where(chipo.item_price == chipo.item_price.max(),chipo.quantity,0)
for i in chipo_expensive:
    if i!=0:
        print("The quantity of the most expensive item ordered is",i)
```

## Step 8. How many times was a Veggie Salad Bowl ordered?

```python
# Checking how many Veggie Salad Bowl ordered
chipo_veggie = np.count_nonzero(chipo.item_name == "Veggie Salad Bowl")
print("Veggie Salad Bowl was ordered", chipo_veggie, "times")
```

# Step 9. How many times did someone order more than one Canned Soda?

```python
# Checking how many times someone order Canned Soda more than 1
chipo_cansoda = np.count_nonzero((chipo.item_name == "Canned Soda") & (chipo.quantity > 1))
print("Canned Soda has been ordered more than once", chipo_cansoda,"times by people")
```
